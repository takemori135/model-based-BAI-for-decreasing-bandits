```python
import matplotlib.pyplot as plt
from nonstationary_bai.plots import *
import seaborn as sns
import pandas as pd


def plot_dim4_and_dim50():
    # plt_expd4 = PltOneGrouofOptArmExp(dim=4, coeffs=[0.5, 0.5, 0.8], n_arms=5, index=[1, 3], sigma=1e-2,)
    # plt_expd50 = PltOneGrouofOptArmExp(dim=50, coeffs=[0.8, 0.2,0.8954915084635513], index=(10, 40), n_arms=5, sigma=1e-2)
    plt.clf()
    # budgets = 50 * np.arange(1, 11)
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 2.5))
    df4 = make_d4_df()
    df50 = make_d50_df()
    sns.pointplot(data=df4, x="budget", y="accuracy", hue="method", errorbar=('ci', 95), ax=ax1, legend=False, linestyles=["solid"]*6 + ["dashed"]*4)
    ax1.set_title("Lower dimension (d=4)")
    sns.pointplot(data=df50, x="budget", y="accuracy", hue="method", errorbar=('ci', 95), ax=ax2, legend="full", linestyles=["solid"]*6 + ["dashed"]*4)
    ax2.set_title("Higher dimension (d=50)")
    sns.move_legend(ax2, "upper left", bbox_to_anchor=(1, 1))
    plt.savefig("images/accuracies_synthetic_d4_d50.pdf", dpi=300, bbox_inches="tight")

plot_dim4_and_dim50()
```
